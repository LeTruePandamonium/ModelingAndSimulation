```python
import pandas as pd

# 0. Load the CSV into a DataFrame
df = pd.read_csv("visa_number_in_japan.csv")

# 1. Inspect which years are in your file
years = df['Year'].unique()
print("Years in dataset:", sorted(years))

# 2. Aggregate total visas issued per actual year
yearly = df.groupby('Year')['Number of issued_numerical'].sum().reset_index()

# 3. Display the result to verify
print(yearly)

# 4. Plot it (bar or line)
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(yearly['Year'], yearly['Number of issued_numerical'], edgecolor='black')
plt.title('Total Visas Issued per Year in Japan')
plt.xlabel('Year')
plt.ylabel('Total Visas Issued')
plt.grid(axis='y')
plt.ticklabel_format(style='plain', axis='y')  # <-- disables scientific notation for y-axis
plt.show()
```
